```
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt  
import io
df = pd.read_csv(io.BytesIO(uploaded['MD_COVID-19.csv']))
df_regions = df[df.columns[2:]]
df_regions
```

```
sum_cases = df_regions.sum(axis = 0, skipna = True)
sum_cases
```

```
regions = list(sum_cases.keys()) 
```

```
values = []
for i in regions:
  values.append(sum_cases[i])
values
```

```
range(int(min(values)), int(max(values)))
```

```
values_new = np.linspace(0, 4850000, num=20)
values_new = values_new.astype(int)
values_new
```

```
fig = plt.figure(figsize = (15, 8)) 
plt.bar(regions, values, width = 0.7) 
  
plt.xlabel("Regions") 
plt.ylabel("No. of cases reported (in millions)") 
plt.title("Total number of cases reported in different regions")
plt.yticks(values_new)
plt.show() 
```

```
fig = plt.figure(figsize =(10, 7)) 
plt.pie(values, labels = regions) 
  
# show plot 
plt.show() 
```
